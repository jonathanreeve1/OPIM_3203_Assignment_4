### Question 4. (12 points + 3 bonus points in total) Aggregating the COVID-19 data at the US county level

Question 4 contains tasks 4a through 4c that you need to complete.

```python
# 4a. (6 points) Write Python code in this code cell
# to use the groupby() method on the DataFrame df2_20211231
# to get information about df2_20211231 at the state level.
# Also print out the total number of records in each state of the DataFrame df2_20211231.

# Also run this code cell after the code is completed.
# Make sure that you execute the code cells above about generating the DataFrame df2_20211231
# before running this code cell.

for group, frame in df2_20211231.groupby('state'):
  num_record = len(frame)
  print(str(num_record) + ' record(s) in group ' + str(group))
```

```python
# 4b. (6 points) Write Python code in this code cell
# to group the DataFrame df2_20211231 by the state name,
# and get the average number of cases of counties for each state in the DataFrame df2_20211231.

# Also run this code cell after the code is completed.
# Make sure that you execute the code cells above about generating the DataFrame df2_20211231
# and importing the libraries such as pandas and numpy
# before running this code cell.

df2_20211231.groupby('state').agg({'cases':[np.average]})
```

```python
# 4c. (3 bonus points) Write Python code in this code cell
# to group the DataFrame df2_20211231 by the state name,
# and get the total number of deaths of counties,
# as well as the average number of deaths of counties,
# for each state in the DataFrame df2_20211231.

# Also run this code cell after the code is completed.
# Make sure that you execute the code cells above about generating the DataFrame df2_20211231
# and importing the libraries such as pandas and numpy
# before running this code cell.

df2_20211231.groupby('state').agg({'cases':[np.sum]})
df2_20211231.groupby('counties').agg({'deaths':[np.average]})
```
